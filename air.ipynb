```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('air.csv', delimiter=';')
```

```python
df
```

```python
def correct_day(date):
    dt = datetime.strptime(date[:10], '%Y-%m-%d')
    return dt.year == 2021 and dt.month == 8 and dt.day == 6
```

```python
validation_col = df['Last Updated'].apply(lambda x: correct_day(x))
```

```python
df = df[validation_col]
```
