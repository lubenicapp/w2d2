```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('invoice.csv', delimiter='|')
```

```python
df.sample(8)
```

```python
def name_cleaner(name):
    return name[name.index('(') + 1:name.index(')')]
```

```python
v_name_cleaner = np.vectorize(name_cleaner)
```

```python
df['SITE_NAME'] = v_name_cleaner(df['SITE_NAME'])
```

```python
df.sample(8)
```

```python
df.drop('INVOICE_NAME', axis=1, inplace = True)
```

```python
df.sample(8)
```

```python
from datetime import datetime
```

```python
july = datetime.strptime('31/07/2018', '%d/%m/%Y')
```

```python
feb = datetime.strptime('10/02/2018', '%d/%m/%Y')
```

```python
(july - feb).days
```

```python
def to_time(string):
    return datetime.strptime(string, '%d/%m/%Y')
```

```python
df['PERIOD'] = (df['END'].apply(lambda x: to_time(x)) - df['BEGIN'].apply(lambda x: to_time(x)))
```

```python
df['PERIOD_IN_DAYS'] = df['PERIOD'].apply(lambda x: int(str(x).split(' ')[0]))
```

```python
df
```

```python
df['KWH_PERIOD'] = df['KWH_PERIOD'].apply(lambda x: float(str(x).replace(' ', '')))
```

```python
df['AVG_KWH_PER_DAY'] = df['KWH_PERIOD'] / df['PERIOD_IN_DAYS']
```

```python
df
```

```python
gb = df.groupby('SITE_NAME').count()
```

```python
def getvalue(x):
    return gb.loc[x]['BEGIN']

df['TOTAL_INVOICE'] = df['SITE_NAME'].map(getvalue)
```
